```python
import niftystocks as ns
from nsepython import equity_history  # Corrected function name
from datetime import datetime, timedelta
import pandas as pd

# 1. Get the list of Nifty 50 stocks
# niftystocks gives us: ['RELIANCE', 'TCS', 'INFY', ...]
stocks = ns.get_nifty50_stocks()
# 2. Set the 1-year date range
# Note: NSE usually expects 'DD-Mon-YYYY' or 'DD-MM-YYYY'
end_date = datetime.now().strftime("%d-%m-%Y")
start_date = (datetime.now() - timedelta(days=365)).strftime("%d-%m-%Y")

# 3. Fetch data for the first stock in the list
symbol = stocks[0]
print(f"Fetching 1 year of data for {symbol}...")

# Use equity_history with the correct parameters
try:
    df = equity_history(symbol, "EQ", start_date, end_date)
    print(df.head())
except Exception as e:
    print(f"Error fetching data: {e}")
```
